```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
train= pd.read_csv('train.csv')

```

```python
train.head()
```

```python
train.info()
```

```python
train['state']=train['state'].astype('category')
train.area_code= train.area_code.astype('category')


```

```python
yes_binary= {'yes':'1','no': '0'}
```

```python
train['international_plan']=train['international_plan'].map(yes_binary).astype('int64')
train['voice_mail_plan']=train['voice_mail_plan'].map(yes_binary).astype('int64')
train['churn']=train['churn'].map(yes_binary).astype('int64')


```

```python
train_dummies= pd.get_dummies(train)
```

```python
len(train_dummies.columns)
```

```python
train_dummies.columns
```

```python
shuffled= train_dummies.sample(frac=1).reset_index().drop('index',axis=1)
```

```python
shuffled_scaled= (shuffled- shuffled.min())/(shuffled.max()-shuffled.min())
```

```python
shuffled_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=8)
pca.fit_transform(shuffled_scaled.drop('churn',axis=1))
```

```python
pca_data= pd.DataFrame(pca.components_,columns=shuffled_scaled.drop('churn',axis=1).columns,index= ['Pca1',
                                                                                         'Pca2',
                                                                                         'Pca3',
                                                                                         'Pca4',
                                                                                         'Pca5',
                                                                                         'Pca6',
                                                                                         'Pca7',
                                                                                         'Pca8'])
```

```python
pca_data
```

```python
pca_Data=pca.fit_transform(shuffled_scaled.drop('churn',axis=1))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.ensemble import BaggingClassifier , GradientBoostingClassifier, AdaBoostClassifier, RandomForestClassifier
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
X_train,X_test,y_train,y_test= train_test_split(pca_Data,shuffled_scaled[['churn']],test_size=0.25)
```

```python
models = []
models.append(('KNN', KNeighborsClassifier()))
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('QDA', QuadraticDiscriminantAnalysis()))
models.append(('Tree', DecisionTreeClassifier()))
models.append(('svc', SVC()))
models.append(('BC', BaggingClassifier(base_estimator=LogisticRegression())))
models.append(('AB', AdaBoostClassifier(base_estimator=LogisticRegression())))
models.append(('GBM', GradientBoostingClassifier()))
models.append(('RF', RandomForestClassifier()))
models.append(('XGB', XGBClassifier()))
```

```python
results= []
names= []
for name, model in models:
    cv_res= cross_val_score(model,X_train,y_train,cv=10,scoring='accuracy',n_jobs=-1)
    results.append(cv_res)
    names.append(name)
    res= '%s: %f. (%f)' %(name,cv_res.mean(),cv_res.std())
    print(res)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
paremeters={ 'learning_rate':[0.1],
            'min_child_weight': [1],
            'max_depth': [9,12],
            'objective': ['binary:logistic'],
            'gamma': np.linspace(0.01,0.9,30)
    
}
```

```python
grid_Search_xgb= GridSearchCV(estimator = XGBClassifier(use_label_encoder=False),param_grid=paremeters,
                             scoring='accuracy',cv= 10,n_jobs=-1)

results_xgb= grid_Search_xgb.fit(X_train,y_train)
```

```python
results_xgb.best_score_
```

```python
prediction_xgb=results_xgb.predict(X_test)
```
