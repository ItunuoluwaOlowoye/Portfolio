# Customer Segmentation
This segmentation model is built based on data provided from survey responses of 100 Walmart customers. This data is fictitious but mimicks actual data. These customers were asked questions about their biodata, and their interactions with Walmart and its competitors over three months.

There are four categories used to build the customer segments for this model:
1. **Behavioural Segmentation**: This is based on the customers' purchasing behaviours. It takes into account how often they shop and how much they spend at Walmart, along with their annual income data. This segmentation can identify customers to be targeted with __*specific marketing campaigns*__ tailored towards their needs.
2. **Competitor Landscape/Loyalty Segmentation**: This is based on whether or not customers prefer to shop at Walmart compared to shopping at competitor stores. In understanding the competitor landscape, we can infer how loyal customers are to the Walmart brand. It takes into account how frequently they shop with Walmart, competitors they also shop from, how much they spend at these stores relative to how much they spend at Walmart. This can help to identify situations where __*customer loyalty*__ needs to be developed and a further analysis can help develop targeted marketing strategies.
3. **Product Preference Segmentation**: This is based on preference for product offerings (such as groceries, clothing items and accessories, homeware, wellness and so on). It accounts for how frequently customers shop for the different products. This segmentation can be used to target customers with __*personalized recommendations*__ tailored towards their needs.
        4. **Channel Preference Segmentation**: This is based on customers preferred mode of shopping, whether instore,
        through the app, or online. Preferences vary for the different product offerings. For example, a customer can
        prefer to buy food instore but buy fashion items online. Therefore, this segmentation is broken down by the
        channel preference for each product. This helps in creating optimization strategies for __*improving customer
        experience*__.
